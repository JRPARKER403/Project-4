```python
#Need to clean this up, she a mess
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import roc_auc_score
import pandas as pd
import numpy as np


from warnings import simplefilter
from pathlib import Path
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn import preprocessing

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Display all of the columns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('Resources/mushrooms.csv')
df
```

```python
df['class'].unique()
```

```python
# Split target column from dataset
y = df['class']
X = df.drop(columns='class')
```

```python
value_counts=df['class'].value_counts()
print(value_counts)
```

```python
colors = ['#81d8d0','#800080']
labels = 'Edible', 'Poisonous'


plt.pie(value_counts,
      labels = labels,
      colors=colors,
      autopct='%1.1f%%',
      startangle=90,
      rotatelabels=True,
      explode=(0.1, 0),
      shadow=True)


plt.title('Mushroom Class Count')
plt.axis('equal')


plt.show()
```

```python
# Encode the categorical variables using get_dummies
X = pd.get_dummies(X)
```

```python
# Preview the data
X.head()
```

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Create the StandardScaler instance
scaler = StandardScaler()
# Fit the Standard Scaler with the training data
X_scaler = scaler.fit(X_train)
# Scale the training data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
model = KNeighborsClassifier(n_neighbors=250)
```

```python
# Train the model
model.fit(X_train_scaled, y_train)
```

```python
# Create predictions
y_pred = model.predict(X_test_scaled)

# Review the predictions
y_pred
```

```python
# Print confusion matrix
cm2=confusion_matrix(y_pred,y_test)
cm2
```

```python
print(classification_report(y_pred,y_test))
```

```python
cm2 = metrics.ConfusionMatrixDisplay(confusion_matrix = cm2, display_labels = [False, True])
cm2.plot()
plt.show()
```

```python
labelEncoder = preprocessing.LabelEncoder()
for col in df.columns:
    df[col] = labelEncoder.fit_transform(df[col])

# Train Test Split
X = df.drop('class', axis=1)
y = df['class']

df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
X_train.shape
```

```python
classifier = LogisticRegression(solver='lbfgs', max_iter=600, random_state=42)
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
# Display the accuracy score for the test dataset.
accuracy_score(y_test, predictions)
```

```python
value_counts5=results['Prediction'].value_counts()
print(value_counts5)
```

```python
value_counts6=results['Actual'].value_counts()
print(value_counts6)
```

```python
colors = ['#81d8d0','#800080']
labels = 'Edible', 'Poisonous'


plt.pie(value_counts5,
      labels = labels,
      colors=colors,
      autopct='%1.1f%%',
      startangle=90,
      rotatelabels=True,
      explode=(0.1, 0),
      shadow=True)


plt.title('Mushroom Class Count Predicted')
plt.axis('equal')


plt.show()
```
